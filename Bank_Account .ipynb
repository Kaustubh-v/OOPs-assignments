```python
import random

class bank_account:
	def __init__(self):
		self.name = " "
		self.type = " "
		self.account_number = 0
		self.balance = -1
 
	def open_account(self):
		self.name = input("Enter account holder's name: ")
		print("Which type of account do you want to open? \n 1. Savings \n 2. Current")
		temp = -1
		while(temp != 1 or temp != 2):
			temp = int(input())
			if(temp == 1):
				self.type = "Savings"
				break
			elif(temp == 2):
				self.type = "Current"
				break
			else:
				print("Enter a valid type!")

        
		while(self.balance < 2500):
			self.balance = int(input("Enter the amount you want to deposit in the account, minimum $2500: "))
			if(self.balance < 2500):
				print("Minimum balance is $2500, may incur a penalty!\n")
        

		if(self.type == "Current"): 
			self.account_number = random.randint(100000,199999)
		else:
			self.account_number = random.randint(200000,299999)
 
		print("Account generation Successful!\n")
            
	def get_account_details(self):
		if(self.name == " "):
			print("Bank account with given name does not exist")
			return
		print("Account Holder: ",self.name,"\n")
		print("Account type :",self.type,"\n")
		print("Account Balance :"," $",self.balance,"\n")
		print("Account number :",self.account_number,"\n")
        
	def deposit(self,amount):
		self.balance += amount
		print("Deposit successful\n")

	def withdraw(self,amount):
		if(amount > self.balance):
			print("Insufficient balance\n")
			return
		else:
			if(self.balance-amount < 2500):
				print("failed!, remaining amount less than the minimum balance required in account\n")
				return
			else:
				self.balance -= amount
                
	def close_account(self):
		self.name = " "
		self.type = " "
		self.account_number = 0
		self.balance = -1

choice = ""
while(choice != "n"):
	print("do you want to initiate the process of opening bank account? (enter y/n):")
	choice = input()
	if(choice == "y"):
		acc1 = bank_account()
		acc1.open_account()
		service = 1
		while(service != 0):
			print("which of the following services you would like to like avail? (0 for quitting) \n 1.get my account details\n 2. withdraw money \n 3. deposit money\n 4. close account")
			service = int(input())
			if(service == 1):
				acc1.get_account_details()
			if(service == 2):
				print("Enter amount to withdraw: ")
				amount = int(input())
				acc1.withdraw(amount)
			if(service == 3):
				print("Enter amount to deposit: ")
				amount = int(input())
				acc1.deposit(amount)
			if(service == 4):
				acc1.close_account()
			else:
				print("invalid option!")
		if(service == 0):
			print("Thank you for opening bank account with us, for any queries, contact nearest branch")
	elif(choice == "n"):
		print("Goodbye!")
	else:
		print("Enter a valid choice!")
```
