```python

while True:
    print("1. Single Inheritance")
    print("2. Multiple Inheritance")
    print("3. Multilevel Inheritance")
    print("4. Exit")

    choice = input("Enter your choice: ")

    if choice == '1':
        class Animal:
            def __init__(self, name):
                self.name = name
        class Bird(Animal):
            def __init__(self, name, can_fly):
                super().__init__(name)
                self.can_fly = can_fly

            def fly(self):
                if self.can_fly:
                    print(f"{self.name} is flying")
                else:
                    print(f"{self.name} can't fly")

        parrot = Bird("Andrew",True)
        parrot.fly()

    elif choice == '2':
        class A:
            def do_something(self):
                print("Doing something in A")

        class B:
            def do_something(self):
                print("Doing something in B")

        class C(B, A):
            pass

        c = C()
        c.do_something()

    elif choice == '3':
        class Animal:
            def __init__(self, name):
                self.name = name
        
        class Mammal(Animal):
            def __init__(self, name, num_legs):
                super().__init__(name)
                self.num_legs = num_legs

            def give_birth(self):
                print(f"{self.name} is giving birth to a baby")

        class Dog(Mammal):
            def __init__(self, name, sound, num_legs, breed):
                super().__init__(name,num_legs)
                self.sound = sound
                self.breed = breed

            def getlegs(self):
                print(self.num_legs)
            
            def getsound(self):
                print(self.sound)
            
        dog = Dog("doggo","bark","4","labrador")
        dog.getlegs()
        dog.getsound()


    elif choice == '4':
        break

    else:
        print("Invalid choice")

```
