```python
# Runtime Polymorphism (Method Overriding)
class Animal:
    def __init__(self, name):
        self.name = name

    def make_sound(self):
        pass

class Dog(Animal):
    def make_sound(self):
        print("Woof")

class Cat(Animal):
    def make_sound(self):
        print("Meow")

def make_animal_sound(animal):
    animal.make_sound()

def method_overriding_demo():
    while True:
        print("1. Make a Dog sound")
        print("2. Make a Cat sound")
        print("3. Exit")

        choice = input("Enter your choice: \n")

        if choice == '1':
            dog = Dog("Rover")
            make_animal_sound(dog)

        elif choice == '2':
            cat = Cat("Whiskers")
            make_animal_sound(cat)

        elif choice == '3':
            break

        else:
            print("Invalid choice\n")


class Rectangle:
    def __init__(self, width, height):
        self.width = width
        self.height = height

    def __add__(self, other):
        if isinstance(other, Rectangle):
            return Rectangle(self.width + other.width, self.height + other.height)
        else:
            return Rectangle(self.width + other, self.height + other)

    def __str__(self):
        return f"Rectangle ({self.width}, {self.height})"

def operator_overloading_demo():
    while True:
        print("1. Add two Rectangles")
        print("2. Add a Rectangle and an integer")
        print("3. Exit")

        choice = input("Enter your choice: ")

        if choice == '1':
            r1 = Rectangle(2, 3)
            r2 = Rectangle(4, 5)
            r3 = r1 + r2
            print(r3)

        elif choice == '2':
            r1 = Rectangle(2, 3)
            r2 = 2
            r3 = r1 + r2
            print(r3)

        elif choice == '3':
            break

        else:
            print("Invalid choice")


# Parametric Polymorphism (Generics)
def reverse(lst):
    lst.reverse()
    return lst

def generics_demo():
    while True:
        print("1. Reverse a list of integers")
        print("2. Reverse a list of strings")
        print("3. Exit")

        choice = input("Enter your choice: ")

        if choice == '1':
            lst = [1, 2, 3, 4, 5]
            print(reverse(lst))

        elif choice == '2':
            lst = ["apple", "banana", "cherry"]
            print(reverse(lst))

        elif choice == '3':
            break

        else:
            print("Invalid choice")


# Menu-Driven Code to Demonstrate All Types of Polymorphism
def main():
    while True:
        print("1. Method Overriding (Runtime Polymorphism)")
        print("2. Operator Overloading (Ad-hoc Polymorphism)")
        print("3. Generics (Parametric Polymorphism)")
        print("4. Exit")

        choice = input("Enter your choice: ")

        if choice == '1':
            method_overriding_demo()

        elif choice == '2':
            operator_overloading_demo()

        elif choice == '3':
            generics_demo()

        elif choice == '4':
            break

        else:
            print("Invalid choice")
main()

```
