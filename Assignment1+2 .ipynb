```python
import math

class Polygon:
    def __init__(self, sides):
        self.sides = sides
        if sides == 3:
            print("It's a Triangle")
        elif sides == 4:
            print("It's a Quadrilateral")
        elif sides == 5:
            print("It's a Polygon")
            
class Triangle(Polygon):
    def __init__(self, side1, side2, side3):
        super().__init__(3)
        self.side1 = side1
        self.side2 = side2
        self.side3 = side3
        
    def change_dimensions(self, side1, side2, side3):
        self.side1 = side1
        self.side2 = side2
        self.side3 = side3
        
    def get_sides(self):
        print("The three sides are:", self.side1, self.side2, self.side3)
        
    def perimeter(self):
        print("Perimeter of this triangle is:", self.side1 + self.side2 + self.side3)
        
    def area(self):
        semi = (self.side1 + self.side2 + self.side3) / 2
        area = math.sqrt(semi * (semi - self.side1) * (semi - self.side2) * (semi - self.side3))
        print("Area of this triangle is:", area)
 

class Square(Polygon):
    def __init__(self, side):
        super().__init__(4)
        self.side = side
        
    def change_dimensions(self, side):
        self.side = side
        
    def get_sides(self):
        print("The four sides are:", self.side, self.side, self.side, self.side)
        
    def perimeter(self):
        print("Perimeter of this square is:", self.side * 4)
        
    def area(self):
        area = self.side * self.side
        print("Area of this square is:", area)
        
        
class Rectangle(Polygon):
    def __init__(self, side1, side2):
        super().__init__(4)
        self.side1 = side1
        self.side2 = side2
        
    def change_dimensions(self, side1, side2):
        self.side1 = side1
        self.side2 = side2
        
    def get_sides(self):
        print("The four sides are:", self.side1, self.side2, self.side1, self.side2)
        
    def perimeter(self):
        perimeter = (self.side1 + self.side2) * 2
        print("Perimeter of this rectangle is:", perimeter)
        
    def area(self):
        area = self.side1 * self.side2
        print("Area of this rectangle is:", area)


def display_menu():
    print("Menu:")
    print("1. Create Triangle")
    print("2. Create Square")
    print("3. Create Rectangle")
    print("0. Exit")


def get_menu_choice():
    while True:
        try:
            choice = int(input("Enter your choice (0-3): "))
            if choice in range(4):
                return choice
            else:
                print("Invalid choice. Please enter a number between 0 and 3.")
        except ValueError:
            print("Invalid choice. Please enter a number.")


def create_triangle():
    while True:
        side1 = float(input("Enter the length of side 1: "))
        side2 = float(input("Enter the length of side 2: "))
        side3 = float(input("Enter the length of side 3: "))
        if(side1 <= 0 or side2 <= 0 or side3 <= 0):
            print("Invalid choice of sides")
        elif(side1 + side2 < side3 or side1 + side3 > side2 or side2 + side3 > side1):
            print("Invalid choice of sides")
        else:
            break
    triangle = Triangle(side1, side2, side3)
    temp = -1
    while(temp != 0):
        print("Enter the service:\n1. Get area \n 2. Get perimeter \n 3. Change Dimensions \n 4. Get sides \n (press 0 to quit)\n")
        temp = int(input())
        if(temp == 1):
            triangle.area()
        elif(temp == 2):
            triangle.perimeter()
        elif(temp == 3):
            side1 = int(input("Enter the new first side: "))
            side2 = int(input("Enter the new first side: "))
            side3 = int(input("Enter the new first side: "))
        elif(temp == 4):
            triangle.get_sides()
        else:
            print("enter a valid choice")

    
    


def create_square():
    while True:
        side = float(input("Enter the length of a side: "))
        if(side <= 0):
            print("invalid choice of side")
        else:
            break

    square = Square(side)
    temp = -1
    while(temp != 0):
        print("Enter the service:\n1. Get area \n 2. Get perimeter \n 3. Change Dimensions \n 4. Get sides \n (press 0 to quit)\n")
        temp = int(input())
        if(temp == 1):
            square.area()
        elif(temp == 2):
            square.perimeter()
        elif(temp == 3):
            side = int(input("Enter the new side: "))
        elif(temp == 4):
            square.get_sides()
        else:
            print("enter a valid choice")



def create_rectangle():
    while True:
        side1 = float(input("Enter the length of side 1: "))
        side2 = float(input("Enter the length of side 2: "))
        if(side1 <= 0 or side2 <= 0):
            print("invalid sides")
        else:
            break
    rectangle = Rectangle(side1, side2)
    temp = -1
    while(temp != 0):
        print("Enter the service:\n1. Get area \n 2. Get perimeter \n 3. Change Dimensions \n 4. Get sides \n (press 0 to quit)\n")
        temp = int(input())
        if(temp == 1):
            rectangle.area()
        elif(temp == 2):
            rectangle.perimeter()
        elif(temp == 3):
            side1 = int(input("Enter the new first side: "))
            side2 = int(input("Enter the new second side: "))
        elif(temp == 4):
            rectangle.get_sides()
        else:
            print("enter a valid choice")




while True:
    display_menu()
    choice = get_menu_choice()
    
    if choice == 0:
        break
    elif choice == 1:
        create_triangle()
    elif choice == 2:
        create_square()
    elif choice == 3:
        create_rectangle()
    else:
        print("Invalid choice.")


```
